# Paleoclimate timeseries with PYLEOCLIM

So what is this shiny tool we have talked about so much in recent years?


- Pyleoclim is a Python package designed for the analysis of paleoclimate data.

- Pyleoclim leverages various data science libraries (numpy, pandas, scikit-learn) for time series analysis, as well as and Matplotlib and Cartopy for the creation of publication-quality figures.

- The package is designed around object-oriented Series, which can be directly manipulated for plotting, spectral and wavelet analysis, and other time series-appropriate operations. 

- Basic familiarity with Python is essential, and many good tutorials exist on the topic.

<div><button> <a href=https://pyleoclim-util.readthedocs.io/en/master/>Pyleoclim<a></button></div>
    
    

## Importing some important modules 

````python
import pandas as pd  # this is for reading text files into tables or DataFrames
import pyleoclim as pyleo  # this is the main workhorse, with all its utilities.
import numpy as np # to manipulate some arrays
import matplotlib.pyplot as plt
````

Several basic plotting styles are available by default with Pyleoclim. Here, let's go with the journal style.

````python
pyleo.set_style('journal_spines_nogrid', font_scale=1)
````

Let's now read the spannagel data from the NOAA website into a `pandas.DataFrame` object.

````python
PATH = 'https://www.ncei.noaa.gov/pub/data/paleo/speleothem/europe/austria/spannagel2005.txt'

spannagel = pd.read_csv(PATH, encoding = 'unicode_escape',
            index_col = 0,
            usecols = [0,1,2,3], 
            header = 43, sep = '\s+', 
            names = ['Ageb2k',   'Cal.Year'  , 'del18O' ,  'Temp °C'])
````

A `pandas.Dataframe` object is simply a object of tabular data, like an excel sheet. It can be indexed or subsetted as you wish. Each column is a `pandas.Series` object.

````python
spannagel.head()
````

````python
spannagel['Temp °C'].head()
````

Having read the data into a dataframe, we are able to creat our main workhorse, the `Pyleo.Series` object.

````python
TIME = spannagel['Cal.Year']
D18O = spannagel['del18O']
TEMP = spannagel['Temp °C']


TS_SPANNAGEL = pyleo.Series(time=TIME, 
                       value=D18O, 
                       value_name='$\delta^{18}$O', 
                       value_unit = '‰ VPDB',
                       time_name='Age', 
                       time_unit='yr C.E')

TS_SPANNAGEL_TEMP = pyleo.Series(time=TIME, 
                       value=TEMP, 
                       value_name='Temperature', 
                       value_unit = '°C',
                       time_name='Age', 
                       time_unit='yr C.E')
````

### Step 2 preprocessing.

Let's look at some preprocessing steps on the data, like for instance, making the timeseries zero-mean, i.e. calculating the anomaly, or making it z-standard (zero mean, unit variance).

#### Anomaly

````python
TS_SPANNAGEL_ANOM = TS_SPANNAGEL.anomaly()

fig, ax = TS_SPANNAGEL.plot(label='original', mute=True, zorder=99)  # use mute=True to hold the showing of the figure
ax = TS_SPANNAGEL_ANOM.plot(label='anomaly', ax=ax, lgd_kwargs={'ncol': 2})
pyleo.showfig(fig)
````

#### Z-standardisation

````python
TS_SPANNAGEL_STD = TS_SPANNAGEL.standardize()

fig, ax = TS_SPANNAGEL.plot(label='original', mute=True, zorder=99)  # use mute=True to hold the showing of the figure
ax = TS_SPANNAGEL_STD.plot(label='standardized', ax=ax, lgd_kwargs={'ncol': 2})
pyleo.showfig(fig)
````

````python
TS_SPANNAGEL_STD.stats()
````

We might even think of changing the time axis, which is extremely simply with pyleoclim. We need to apply the method `.convert_time_unit()` to our series.

````python
TS_SPANNAGEL_STD_CE = TS_SPANNAGEL_STD.convert_time_unit('year C.E')

fig, ax = TS_SPANNAGEL_STD_CE.plot() 
````

### Step 3. Power Spectral Density

We show how to chain methods too. We apply the `spectral()` method to the `pyleoclim.Series` object, and then apply the `.plot()` method to this result. By default, the Wavelet Weighted Z-transform method is used.

````python
fig, ax = TS_SPANNAGEL_STD.spectral().plot(label='WWZ')
````

To make sense of the result, namely to identify the periodocity of our data, we need to perform some significance test. 

Currently, _Pyleoclim_ supports the test with AR(1) surrogates, and we can perform the test by simply calling the `signif_test()` method.

Here demonstrate the one liner again, which uses several methods:
```
.standardize()
.convert_time_unit()
.spectral()
.signif_test()
.plot()
```

````python
fig, ax = TS_SPANNAGEL.standardize().convert_time_unit('year C.E').spectral().signif_test(qs = [0.95],number = 400).plot()
````

## Filtering

Filtering can be done using the `.filter()` method, on a regularly space dataset.  Here we can demonstrate how this will work with the spannagel temperature reconstruction, using a cutoff frequency of 1/60 and a **butterworth** filter (by default). 

````python
# first we read the data
TIME = spannagel['AgeBP']
VALUE = spannagel['Temp °C']

TS_SPANNAGEL_TEMP = pyleo.Series(time = TIME, value = VALUE, value_name='Temp', value_unit='°C', time_unit = 'yr BP')
````

````python
# then we convert the time to years CE.
TS_SPANNAGEL_TEMP_CE = TS_SPANNAGEL_TEMP.convert_time_unit('year C.E')
````

````python
fig_filter, ax = TS_SPANNAGEL_TEMP_CE.plot(mute = True, plot_kwargs= {'alpha':0.5, 'lw':0.85}, color = "C3" )

#ax = TS_SPANNAGEL_STD.interp(step=1, method = 'linear').plot(ax = ax,plot_kwargs= {'alpha':0.5, 'lw' : 0.5 },color = "C1")

ax = TS_SPANNAGEL_TEMP_CE.interp(step=1, method = 'linear').filter(cutoff_freq=1/60).plot(ax = ax ,color = "C3" )

pyleo.showfig(fig_filter)
````

### Step 2. Making the tables into Series objects

We extract the time and value axes from those tables

````python
# The detrended record
D18O = hulu_detrended['Dd18Onrm']
TIME = hulu_detrended['age_calkaBP']

TS_HULU_DETRENDED = pyleo.Series(time = TIME, time_unit='kyr BP',value = D18O,value_name='$\delta^{18}$O$_{nrm}$')
````

````python
# The del18O composite record
D18O = hulu['d18OcarbVPDB']
TIME = hulu['age_calkaBP']

TS_HULU = pyleo.Series(time = TIME, time_unit='kyr BP',value = D18O,value_name='$\delta^{18}$O',value_unit='‰ VPDB')
````

### Step 3. Use another package for daily insolation parameters

````python
import numpy as np
from climlab.solar.orbital import OrbitalTable
from climlab.solar.insolation import daily_insolation
````

Let's make a record of daily insolation on 21st of July, at 65°N.

````python
# array with specified kyears (can be plain numpy or xarray.DataArray)
years = np.linspace(-640, 0, (640*5)+1) # every 200 years

# subset of orbital parameters for specified time
orb = OrbitalTable.interp(kyear=years)

# insolation values for past 5 Myears at 65N at summer solstice (day 172)
S65 = daily_insolation(lat=65, day=172, orb=orb)
````

Now we read the insolation values into a `pyleoclim.Series` object.

````python
INSOL = S65.values
TIME = S65.kyear

TS_INSOL = pyleo.Series( value = -INSOL, value_name='21st July insolation 65°N', value_unit = 'W.m$^{-2}$', time = -TIME, time_unit = 'ka')
````

We can now use the `pyleoclim.MultipleSeries` object.

````python
SPELEOTHEMS = pyleo.MultipleSeries([TS_INSOL, TS_HULU, TS_HULU_DETRENDED], time_unit = 'kyr BP')
````

And now we can make a stacked plot of this SPELEOTHEM `MultipleSeries` object.

````python
fig, ax = SPELEOTHEMS.stackplot()
````

## Spectral analysis of the Speleothem records

In particular, both the detrended (Orbital forcing subtracted) and original d18O composite records of the paper

Notice that we chain methods again; this time, interpolating to 200 year regular intervals with `interp()` and standardizing with `standardize()`. We then perform the`spectral()` and `signif_test()` methods before using `plot()`.

````python
TS_HULU.interp(step = 0.2).standardize().spectral().signif_test(qs = [0.99], number = 200).plot()
````

Naturally several algorithms to perform the spectral analysis are available to you, namely: WWZ, MTM, Lomb-scargle.

````python
psd_ls  = TS_HULU.interp(step = 0.2).standardize().spectral(method='lomb_scargle')
psd_mtm = TS_HULU.interp(step = 0.2).standardize().spectral(method='mtm')
psd_wwz =TS_HULU.interp(step = 0.2).standardize().spectral(method = 'wwz')

fig, ax = psd_mtm.plot(label='MTM',mute=True)
psd_wwz.plot(ax=ax,label='WWZ',mute=True)
psd_ls.plot(ax=ax,label='Lomb-Scargle')
pyleo.showfig(fig)
````

In fact, here we see that MTM produces very broad peaks at orbital periods compared to WWZ's sharp peaks. Lomb-Scargle is very noisy at small frequencies.

````python
psd_wwz_detrend = TS_HULU_DETRENDED.interp(step = 0.2).standardize().spectral()
psd_wwz = TS_HULU.interp(step = 0.2).standardize().spectral()
````

````python
psd_wwz_detrend_sig = psd_wwz_detrend.signif_test(qs = [0.99], number = 400)
psd_wwz_sig = psd_wwz.signif_test(qs = [0.99], number = 400)
````

````python
fig = plt.figure(figsize = (12,5))

ax = fig.add_subplot(121)
psd_wwz_sig.plot(ax = ax, label = 'detrended')

ax2 = fig.add_subplot(122)
ax2 = psd_wwz_detrend_sig.plot(ax = ax2, label = 'composite')

pyleo.showfig(fig)
````

````python
TS_HULU_DETRENDED.interp(step = 0.2).summary_plot(psd=psd_wwz_detrend_sig,n_signif_test=10)
````

````python
TS_HULU.interp(step = 0.2).summary_plot(psd=psd_wwz_sig,n_signif_test=10)
````
